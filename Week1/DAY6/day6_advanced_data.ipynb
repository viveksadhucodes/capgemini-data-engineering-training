### DATA CLEANING (PHASE 1)

```python
# Replace null names and cities with "Unknown"
customers_df = customers.fillna({
    "name": "Unknown Customer",
    "city": "Unknown City"
})
# For the Cars Table
cars_df = cars.fillna({
    "brand": "Generic",
    "model": "Standard",
    "price": 0
})
sales_df = sales.fillna({
    "sale_date": "2024-01-01" # Or a default date
})
```

```python
#removing negative values 
from pyspark.sql.functions import abs, col

# Apply to the price column
cars_df = cars_df.withColumn("price", abs(col("price")))
cars_df.show()

#validating
print("Remaining negatives: ",cars_df.filter(col("price")<0).count())
```

```python
# trim strings 
from pyspark.sql.functions import trim, col

# Apply trim to the 'brand' column
cars_df = cars_df.withColumn("brand", trim(col("brand")))
```

```python
# remove dulplicates 
sales_final = sales_df.dropDuplicates(["sale_id"])
```

### Quality Control (Phase 2)

```python
invalid_sales_customers = sales_df.join(customers_df, "customer_id", "left_anti") # left_anti returns rows that do not have a match in the right table 
invalid_sales_cars = sales_df.join(cars_df, "car_id", "left_anti")
print("Number of sales with invalid customer links:", invalid_sales_customers.count())
```

```python
from pyspark.sql.functions import col

# 1. Total Raw Counts
raw_sales_count = sales_df.count()

# 2. Count Nulls (example: missing price)
null_price_count = cars_df.filter(col("price").isNull()).count()

# 3. Count Duplicates (based on sale_id)
duplicate_count = sales_df.count() - sales_df.dropDuplicates(["sale_id"]).count()

# 4. Count Orphans (Invalid Foreign Keys using Left Anti Join)
# Finding sales that don't have a matching car in the cars_df
orphan_cars_count = sales_df.join(cars_df, "car_id", "left_anti").count()

# 5. Final Clean Count (assuming df_final is your joined, cleaned result)
final_count = df.count()
```

```python
validation_data = [
    ("Total Raw Records", raw_sales_count),
    ("Null Price Records", null_price_count),
    ("Duplicate Sale IDs", duplicate_count),
    ("Orphan Records (Invalid Car IDs)", orphan_cars_count),
    ("Total Clean Records for Analysis", final_count)
]

# Create Schema
schema = ["Validation_Check", "Record_Count"]

# Create the Report DataFrame
validation_report_df = spark.createDataFrame(validation_data, schema)

# Show the report
validation_report_df.show(truncate=False)
```

```python
from pyspark.sql.functions import lit

validation_report_df.withColumn("Percentage", (col("Record_Count") / raw_sales_count) * 100).show()
```

### TRANSFORMATIONS (Phase 3)


```python
# Join the tables, but rename the dealer name to avoid the conflict
df = sales.join(customers, "customer_id") \
          .join(cars, "car_id") \
          .join(sales_dealer, "sale_id") \
          .join(dealers.withColumnRenamed("name", "dealer_name"), "dealer_id")

# Now your transformation cell will work because 'name' only refers to the Customer
customer_revenue = df.groupBy("customer_id", "name") \
    .agg(sum("price").alias("total_spent")) \
    .orderBy(desc("total_spent"))

customer_revenue.show()
```

```python
#brand-wise sales
cars_per_brand=df.groupBy("brand").agg(countDistinct("car_id").alias("cars_sold")).orderBy(desc("cars_sold"))
cars_per_brand.show()
```

```python
# Rename 'city' and 'name' in the Dealers table to avoid conflicts
df = sales.join(customers, "customer_id") \
          .join(cars, "car_id") \
          .join(sales_dealer, "sale_id") \
          .join(dealers.withColumnRenamed("city", "dealer_city") \
                       .withColumnRenamed("name", "dealer_name"), "dealer_id")

# Now your city-wise sales code will work perfectly!
city_revenue = df.groupBy("city") \
    .agg(sum("price").alias("city_total_revenue")) \
    .orderBy(desc("city_total_revenue"))

city_revenue.display()
```

### DEALER ANALYTICS (PHASSE 4)

```python
# revenue per dealer 
from pyspark.sql.functions import sum, desc

dealer_revenue = df.groupBy("dealer_id", "dealer_name") \
    .agg(sum("price").alias("total_revenue")) \
    .orderBy("dealer_id")

dealer_revenue.show()
```

```python
#top 5 dealers 
top_dealers = df.groupBy("dealer_name") \
    .agg(sum("price").alias("total_revenue")) \
    .orderBy(desc("total_revenue"))

top_dealers.show(5) 
```

```python
#dealer-citywise revenue 
dealer_city_contribution = df.groupBy("dealer_city") \
    .agg(sum("price").alias("showroom_revenue")) \
    .orderBy(desc("showroom_revenue"))

dealer_city_contribution.display()
```

### SQL ANALYSIS


```python
# create a temp view
df.createOrReplaceTempView("final_sales_table")
```

```python

%sql
SELECT city, name, total_spent
FROM (
  SELECT 
    city, 
    name, 
    SUM(price) as total_spent,
    DENSE_RANK() OVER (PARTITION BY city ORDER BY SUM(price) DESC) as rank
  FROM final_sales_table
  GROUP BY city, name
)
WHERE rank <= 3
ORDER BY city, rank;
```

```python

%sql
SELECT 
    TRUNC(sale_date, 'MM') AS sale_month,
    COUNT(sale_id) AS total_units_sold,
    SUM(price) AS total_revenue
FROM final_sales_table
GROUP BY sale_month
ORDER BY sale_month;
```

```python

%sql
SELECT 
    customer_id,
    name, 
    COUNT(sale_id) AS purchase_count,
    SUM(price) AS lifetime_value
FROM final_sales_table
GROUP BY customer_id, name
HAVING COUNT(sale_id) > 1
ORDER BY purchase_count DESC;
```
