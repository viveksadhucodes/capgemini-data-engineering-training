```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("E2E_Pipeline").getOrCreate()
data = [
(1, "C001", "Laptop", "50000", "2024-01-01"),
(2, "C002", "Mobile", None, "2024-01-02"),
(3, "C003", "Tablet", "20000", "2024-01-03"),
(4, "C004", "Laptop", "55000", "2024-01-04"),
(5, "C005", "Headphones", None, "2024-01-05")
## Dont use this for first time (3, "C003", "Tablet", "22000", "2024-01-06"), # updated
# (6, "C006", "Camera", "30000", "2024-01-06"),
# (7, "C007", "Mobile", "18000", "2024-01-07"),
# (8, "C008", "Watch", "8000", "2024-01-07")
]
columns = ["order_id", "customer_id", "product", "amount", "updated_date"]
df = spark.createDataFrame(data, columns)
```

```python
# Handle null
const_val = 1000

# Use a dictionary: {"column_name": value}
df_clean = df.fillna({"amount": const_val})

df_clean.display()
```

```python
#Task 2: Cast Columns Convert amount → integer Convert updated_date → date
from pyspark.sql.functions import *

df = df.withColumn("amount", df.amount.cast("int"))
df.printSchema()
```

```python
# Task 3: Add Derived Columns
# bonus = amount * 0.1
# category:
# 20000 → High
# else → Low

from pyspark.sql.functions import *

df = df.withColumn("bonus", df.amount * 0.1)
df = df.withColumn("category", when(df.amount >= 20000, "High").when(df.amount < 20000, "Low").otherwise("Unknown"))
df.display()

```

```python
def amount_bucket(amount):
    if amount is None:
        return "Unknown"
    if amount < 10000:
        return "Low"
    elif amount >= 10000 and amount <= 30000:
        return "Medium"
    else:
        return "High"

```

```python
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType

# 1. Register the UDF with an explicit return type
amount_bucket_udf = udf(lambda amount: amount_bucket(amount), StringType())

# 2. Cast the column to Integer FIRST, then apply the UDF
# This fixes the 'str' vs 'int' comparison error
df = df_clean.withColumn("amount", col("amount").cast("int")) \
             .withColumn("amount_bucket", amount_bucket_udf(col("amount")))

df.display()
```

```python
# Task 5: Rename Columns
df = df.withColumnRenamed("updated_date", "updated")
df = df.withColumnRenamed("amount", "price")
df = df.withColumnRenamed("bonus", " bonus")
display(df)
```

```python
data_day2=[
(6, "C006", "Camera", "30000", "2024-01-06"),
(7, "C007", "Mobile", "18000", "2024-01-07"),
(8, "C008", "Watch", "8000", "2024-01-07")
]
df_day2 = spark.createDataFrame(data_day2, columns)
df_day2.display()
```

```python
df_day2= df_day2.withColumnRenamed("updated_date", "updated")
amount_bucket_udf = udf(lambda amount: amount_bucket(amount), StringType())

# 2. Cast the column to Integer FIRST, then apply the UDF
# This fixes the 'str' vs 'int' comparison error
df_day2= df_day2.withColumn("amount", col("amount").cast("int")) \
             .withColumn("amount_bucket", amount_bucket_udf(col("amount")))
```

```python
df_day2.display()
```

```python
filtered_existing = df.join(
    df_day2,
    on="order_id",
    how="left_anti"
    
)
```

```python
final_df = filtered_existing.union(df)
final_df.display()
```
