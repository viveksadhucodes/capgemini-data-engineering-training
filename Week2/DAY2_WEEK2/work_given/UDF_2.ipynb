### WHY DO WE NEED UDF?

### Can we write every logic using built-in functions?

### Example Problem

```python
If salary is NULL → "Unknown"
If salary > 6000 → "A"
Else → "B"
Too complex
Not directly available in PySpark functions
```

### WHAT IS UDF?
### UDF = User Defined Function
### It means:
### “You write your own Python function and apply it on DataFrame columns”

### Create normal Python function

```python
def grade(salary):
    if salary is None:
        return "Unknown"
    elif salary > 6000:
        return "A"
    else:
        return "B"
```

### Convert to UDF

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

grade_udf = udf(grade, StringType())
```

### Spark doesn’t understand Python directly
### So we “wrap” function into UDF
### Must specify return type

### Apply on DataFrame

```python
df = df.withColumn("grade", grade_udf(df.salary))
df.display()
```
