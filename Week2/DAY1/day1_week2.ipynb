### Cleaning


```python
from pyspark.sql.functions import col, abs, trim, lower, coalesce, lit

# 1. Correct Negative Values
# Using abs() ensures we keep the magnitude of the transaction but remove the sign error
policies_cleaned = policies.withColumn("premium", abs(col("premium")))
claims_cleaned = claims.withColumn("claim_amount", abs(col("claim_amount")))

# 2. Trim and Standardize Strings
# Standardizing 'status' and 'city' helps with consistent grouping/filtering
customers_cleaned = customers.withColumn("city", trim(col("city"))) \
                             .withColumn("name", trim(col("name")))

claims_cleaned = claims_cleaned.withColumn("status", trim(col("status")))

# 3. Handle Null Values
# Fill missing premiums with 0 and missing claim amounts with 0 to avoid NULL math issues
policies_cleaned = policies_cleaned.fillna({"premium": 0})
claims_cleaned = claims_cleaned.fillna({"claim_amount": 0, "status": "Unknown"})

# 4. Re-run Join for Cleaned Data
df_cleaned = customers_cleaned.join(policies_cleaned, "customer_id", "left") \
                              .join(claims_cleaned, "policy_id", "left") \
                              .join(policy_agent, "policy_id", "left") \
                              .join(agents, "agent_id", "left")

# View cleaned result
df_cleaned.select("customer_id", "premium", "claim_amount", "city", "status").show(10)
```

### Validation


```python
# 1. Identify Invalid Foreign Keys (Orphans) using Left-Anti Joins
# Policies pointing to non-existent customers
invalid_policies = policies.join(customers, "customer_id", "left_anti")

# Claims pointing to non-existent policies
invalid_claims = claims.join(policies, "policy_id", "left_anti")

# Policy-Agent mappings with missing agents or policies
invalid_policy_agents = policy_agent.join(policies, "policy_id", "left_anti") \
                                    .union(policy_agent.join(agents, "agent_id", "left_anti"))

# 2. Row Count Checks
# Comparing raw data vs. joined result to detect cartesian products or loss
count_customers = customers.count()
count_policies = policies.count()
count_claims = claims.count()
count_final = df_cleaned.count()

# 3. Validation Report Generation
validation_summary = spark.createDataFrame([
    ("Customers", count_customers, 0, count_customers),
    ("Policies", count_policies, invalid_policies.count(), policies_cleaned.count()),
    ("Claims", count_claims, invalid_claims.count(), claims_cleaned.count()),
    ("Final Joined Records", "N/A", "N/A", count_final)
], ["Entity", "Total Raw Records", "Invalid (Orphans)", "Cleaned Records"])

validation_summary.show()

# Optional: Preview invalid records for debugging
print("Sample of Policies with missing Customer IDs:")
invalid_policies.select("policy_id", "customer_id").show(5)
```

### Transformations

```python
from pyspark.sql.functions import sum, col, round, when, lit

# 1. Customer-Level Aggregation
# We use customers_cleaned.name specifically to avoid ambiguity with agents.name
customer_metrics = df_cleaned.groupBy("customer_id", customers_cleaned.name, "city") \
    .agg(
        sum("premium").alias("total_premium"),
        sum("claim_amount").alias("total_claim")
    )

# 2. Compute Risk Score
customer_metrics = customer_metrics.withColumn(
    "risk_score", 
    when(col("total_premium") > 0, round(col("total_claim") / col("total_premium"), 2))
    .otherwise(lit(0))
)

# 3. City-Wise Distribution
city_distribution = customer_metrics.groupBy("city") \
    .agg(
        sum("total_premium").alias("city_total_premium"),
        sum("total_claim").alias("city_total_claim"),
        round(sum("total_claim") / sum("total_premium"), 2).alias("city_risk_index")
    )

# --- View Results ---

print("Customer Risk Profile:")
# Note: Using col("name") here is safe because the groupBy narrowed it down to one "name" column
customer_metrics.select("customer_id", "name", "total_premium", "total_claim", "risk_score") \
    .orderBy(col("risk_score").desc()).show(10)

# Databricks specific display
display(customer_metrics.orderBy(col("risk_score").desc()))

print("City-Wise Insurance Distribution:")
city_distribution.orderBy(col("city_total_premium").desc()).show()
```
